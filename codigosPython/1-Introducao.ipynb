Algoritmos de trocar uma lmpada

```python
def trocar():
    print('Pegar escada e colocar embaixo do bocal')
    print('Pegar uma lmpada nova')
    print('Subir na escada at alcanar o bocal')
    print('Tirar a lmpada velha do bocal')
    print('Colocar a lmpada nova no bocal')
    print('Descer da escada')
    print('Guardar a escada')
    print('Descartar adequadamente a lampada velha')
trocar()
```

Foi trocado a lmpada sem testar se era realmente necessario

```python
def verificacao():
    print('Ligar o interruptor')
    print('Se a lmpada no acender, ento')
    trocar()
verificacao()
```
