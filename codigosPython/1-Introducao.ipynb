Algoritmos de trocar uma lâmpada

```python
def trocar():
    print('Pegar escada e colocar embaixo do bocal')
    print('Pegar uma lâmpada nova')
    print('Subir na escada até alcançar o bocal')
    print('Tirar a lâmpada velha do bocal')
    print('Colocar a lâmpada nova no bocal')
    print('Descer da escada')
    print('Guardar a escada')
    print('Descartar adequadamente a lampada velha')
trocar()
```

Foi trocado a lâmpada sem testar se era realmente necessario

```python
def verificacao():
    print('Ligar o interruptor')
    print('Se a lâmpada não acender, então')
    trocar()
verificacao()
```

... e se mesmo após a troca a lâmpada não acender??? É preciso testar até que atenda o problema e se o problema for no bocal ou no interruptor?

```python
def guardar():
    print('Descer da escada')
    print('Guardar a escada')
    print('Descartar adequadamente a lampada velha')
```

```python
#A lâmpada irá funcionar na 4 troca 
lampada = 0
for i in range(5):
    if lampada == 0:
        verificacao()
    else:
        guardar()
    if i == 4:
        lampada=1
```

```python
lampada = 0
def ligarInterruptor(lampada):
    if lampada == 0:
        print('Pegar uma escada e posicionar embaixo da lâmpada')
        print('Buscar lampadas reservas')
        print('Subir a escada até a altura de troca')
        for i in range(5):
            if i == 1:
                lampada = 1
                print('Funcionou')
                print(f'Descer da escada\nGuardar a escada(ou deslocar ao proximo soquete)\nDescartar as lâmpadas que não funcionaram\nGuardar as lâmpadas reservas que sobraram')
            if lampada == 0:
                print('Retirar a lâmpada queimada')
                print('Colocar a lâmpada reserva')
            
ligarInterruptor(lampada)
```

Vamos imaginar que temos 5 soquetes a serem testados, podemos aproveitar a função acima
